## Notebook demonstrating getting PDB file contents with Python urlopen

Adapted from `urlopen` use in 'Karsten's program to measure Cbeta distances.py' file.

```python
import urllib.request
pdbfile = urllib.request.urlopen("http://www.pdb.org/pdb/files/1mvp.pdb")
for line in pdbfile:
    print(line)
```

```python
import urllib.request
pdbfile = urllib.request.urlopen("http://www.pdb.org/pdb/files/1mvp.pdb")
for line in pdbfile:
    print(line.decode())
```

```python
import urllib.request
pdbfile = urllib.request.urlopen("http://www.pdb.org/pdb/files/1mvp.pdb")
for line in pdbfile:
    print(line.strip().decode())
```
